# Bag of Words

Bag of Words (ou saco de palavras) é um método de representação computacional de um texto baseado puramente no léxico, i.e. sem se preocupar com informações estruturais ou retóricas. Nele um texto é representado pela contagem das palavras que o compõem.

```python
# O SciKit-Learn provê uma classe para realização
from sklearn.feature_extraction.text import CountVectorizer
bag_of_words = CountVectorizer()


texts = [corpus.raw(f) for f in corpus.fileids()]
X =  bag_of_words.fit_transform(texts).toarray()


print(bag_of_words.get_feature_names())
print(len(bag_of_words.get_feature_names()))
print(X)
```

Essa representação é útil para tarefas que são majoritariamente dependentes do léxico. Por exemplo, a categorização de texto por tópicos.

```python
from sklearn import svm

clf = svm.SVC()
y = [corpus.categories(f)[0] for f in corpus.fileids()]
train = [(X[i],y[i]) for i in range(len(X))]
random.shuffle(train)
X = [t[0] for t in train]
y = [t[1] for t in train]
clf.fit(X[:-10],y[:-10])

predictions = clf.predict(X[-10:])
for i in range(10):
  print("Categ:",y[-10+i],"-> Previsto:",predictions[i])
```

# TF-IDF
TF-IDF, ou Term Frequence - Inverse Document Frequency, é uma métrica para descriver a descriptividade de um termo em relação a um documento. Ela é similar à medida normalizada que fizemos anteriormente sobre as categorias, mas pensadas em termos de cada documento do corpus

A métrica TF-IDF é composta pela multificação de dois fatores:


1.   A Frequencia do Termo: mede a frequência com que um termo aparece num documento
\begin{equation}
tf(w_i,d_j) = \dfrac{w_{ij}}{\sum_{k \in V} w_{kj}}
\end{equation}
com $w_{ij}$ representando a quantidade de vezes que a palavra $w_i$ aparece no documento $d_j$

2.   A frequencia inversa do documento: mede o quão descritiva uma palavra é de um documento em particular, i.e. o quão exclusiva ela é daquele documento. ÉÉ calculada como:
\begin{equation}
idf(w_i,d_j) = log_e(\dfrac{N}{|\{d~|~ w_i \in d\}|})
\end{equation}
em que $N$ é a quantidade de documentos no corpus.





```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
X = tfidf_vectorizer.fit_transform(texts).toarray()
print(tfidf_vectorizer.get_feature_names())
print(len(tfidf_vectorizer.get_feature_names()))
print(X)
```

Podemos também utilizar a métrica TF-IDF para obter representações de um texto baseado em léxico.

```python
clf = svm.SVC()
y = [corpus.categories(f)[0] for f in corpus.fileids()]
train = [(X[i],y[i]) for i in range(len(X))]
random.shuffle(train)
X = [t[0] for t in train]
y = [t[1] for t in train]
clf.fit(X[:-10],y[:-10])

predictions = clf.predict(X[-10:])
for i in range(10):
  print("Categ:",y[-10+i],"-> Previsto:",predictions[i])
```

# N-Gramas

```python
palavras = [w.lower() for w in corpus.words()]
bigrams = [tuple(palavras[i:i+2]) for i in range(len(palavras)) if palavras[i].lower() not in stop_words]
fd = FreqDist(bigrams)
for i in range(0,50):
  print(fd.most_common()[i])
```

```python
dist = {k[0]+' '+k[1]:fd[k] for k in fd.keys() if k[0] not in stop_words}
wordcloud = WordCloud(width = 800, height = 800, background_color = 'black',
                      stopwords = stop_words, max_words = 1000,min_font_size = 20).generate_from_frequencies(dist)

import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["figure.figsize"] = (10,5)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
