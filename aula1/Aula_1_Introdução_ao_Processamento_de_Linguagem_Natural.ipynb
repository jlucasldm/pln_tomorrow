```python
def encode(example):
     text = example['text']
     words = nltk.word_tokenize(text,language="portuguese")
     filtered_sentence = [w for w in words if not w.lower() in stop_words]
     filtered_sentence = [re.sub(r'[^\w\s]', '', token) for token in filtered_sentence if re.sub(r'[^\w\s]', '', token)]
     return {'words':filtered_sentence}
```

```python
train = corpus['train'].map(encode)
print(train)
```

```python
# Como os textos podem ser agrupados por tema?

def freq_texto(texto):
  word_count = dict.fromkeys(texto, 0)
  for word in texto:
    word_count[word] = word_count[word] + 1
  return word_count

def freq_geral(textos):
  all_words = []
  for texto in textos:
    all_words += texto['words']
  return freq_texto(all_words)

print(freq_geral(train))

freq_g = freq_geral(train)
output = dict(sorted(freq_g.items(), key=lambda item: item[1], reverse=True))

print(output)

topicos = {
    'negocios': ['empresas', 'empresa', 'produtores', 'negócios', 'produtor'],
    'mercado': ['preço', 'indústria', 'mrecado', 'preços'],
    'governo': ['governo', 'PMDB', 'ministro'],
    'esporte': ['time', 'Fifa', 'jogos']
}

agrupamento = []

count = 0

for texto in train:
  topicos_palavra = []
  for palavra in texto['words']:
    if topicos['negocios'].count(palavra):
      topicos_palavra.append('negocios')
    if topicos['mercado'].count(palavra):
      topicos_palavra.append('mercado')
    if topicos['governo'].count(palavra):
      topicos_palavra.append('governo')
    if topicos['esporte'].count(palavra):
      topicos_palavra.append('esporte')
  print('Texto ', count, ': ', set(topicos_palavra))
  count += 1
```

```python
# @title
from nltk.corpus.reader.tagged import *
from nltk.tokenize import *
from nltk.corpus.reader.util import *
import re
class ChunkIdentifier:
   def __init__(self,tokenizer=None):
        if not tokenizer:
           self.__tokenizer = WhitespaceTokenizer(pattern='\\s+', gaps=True, discard_empty=True, flags=re.UNICODE|re.MULTILINE|re.DOTALL)
        else:
           self.__tokenizer = tokenizer
   def tokenize(self,sent_str):
      sent_uncat = self.__tokenizer.tokenize(sent_str)
      sent = []
      categ= 'O'
      for word in sent_uncat:
         if re.match("\(\w+",word):
            categ= "B-{}".format(re.match("\((\w+)",word)[1])
            continue
         elif word ==")":
            categ = 'O'
            continue
         else:
            if categ[0]=='B':
               sent.append(word+'/'+categ)
               categ = 'I'+categ[1:]
            elif categ !="O":
               sent.append(word+'/'+categ)
            else:
               sent.append(word+'/O')
      return sent


class ChunkerTokenizer:
   def __init__(self,tokenizer=None):
        if not tokenizer:
           self.__tokenizer = WhitespaceTokenizer(pattern='\\s+', gaps=True, discard_empty=True, flags=re.UNICODE|re.MULTILINE|re.DOTALL)
        else:
           self.__tokenizer = tokenizer
   def tokenize(self,sent_str):
      return [word for word in self.__tokenizer.tokenize(sent_str) if not re.match("[()]",word)]


class ChunkedCorpus(TaggedCorpusReader):
   def __init__(
        self,
        root,
        fileids,
        word_tokenizer=WhitespaceTokenizer(),
        sent_tokenizer=RegexpTokenizer(pattern='\n', gaps=True, discard_empty=True, flags=re.UNICODE|re.MULTILINE|re.DOTALL),
        para_block_reader=read_blankline_block,
        encoding='utf8',
        sep="",
        tagset=None
   ):
        TaggedCorpusReader.__init__(
            self,
            root,
            fileids,
            sep=sep,
            encoding=encoding,
            word_tokenizer = ChunkerTokenizer(word_tokenizer),
            sent_tokenizer = sent_tokenizer,
            para_block_reader = para_block_reader,
            tagset = tagset
        )
        self.__chunks_tokenizer = word_tokenizer
   def chunks(self, fileids=None):
        """
        :return: the given file(s) as a list of
            sentences or utterances, each encoded as a list of word
            strings.
        :rtype: list(list(str))
        """
        return concat(
            [
                [(text,categ) for (fileid,enc) in self.abspaths(fileids, True) for categ,text in re.findall("\((\S+) ([^)]+) \)",self.raw(fileid))]

            ]
        )
   def BIO_words(self, fileids=None):
               return concat(
            [
                TaggedCorpusView(
                    fileid,
                    enc,
                    True,
                    False,
                    False,
                    '/',
                    ChunkIdentifier(self.__chunks_tokenizer),
                    self._sent_tokenizer,
                    self._para_block_reader,
                    None,
                )
                for (fileid, enc) in self.abspaths(fileids, True)
            ]
        )
   def BIO_sents(self, fileids=None):
               return concat(
            [
                TaggedCorpusView(
                    fileid,
                    enc,
                    True,
                    True,
                    False,
                    '/',
                    ChunkIdentifier(self.__chunks_tokenizer),
                    self._sent_tokenizer,
                    self._para_block_reader,
                    None,
                )
                for (fileid, enc) in self.abspaths(fileids, True)
            ]
        )
```

## Identificando informação crucial em um conjunto textual.

Dado um texto, como podemos gerar uma versão condesada da informação expressa num texto, de forma a facilitar a leitura de uma grande coleção de textos/notícias de forma a permitir somente as partes relevantes?


```python
!wget http://marlovss.work.gd:8080/tomorrow/aula1/summ.zip
!unzip summ.zip -d summ
import re,os
for name in os.listdir("summ/"):
     with open(f"summ/{name}") as f:
         text = f.read()
     text = re.sub("\r","",text)
     text = re.sub("\n"," ", text)
     with open(f"summ/{name}",'w') as f:
         f.write(text)
```

```python
import glob
from datasets import load_dataset
files = glob.glob("summ/*.txt")
corpus = load_dataset("text",data_files=files)['train']
```

```python
import nltk
def encode (example):
   text = example['text']
   sentences = nltk.sent_tokenize(text,language='portuguese')
   words = [nltk.word_tokenize(sent, language='portuguese') for sent in sentences ]
   return {'sents':sentences,'words':words}
```

```python
corpus = corpus.map(encode)
```
